```python
lista_vendas = [1000,500,800,1500,200,2300]
meta = 1200
percentual_bonus = 0.1
for venda in lista_vendas:
  if venda >= meta:
    bonus = percentual_bonus * venda
  else:
    bonus = 0
  print(bonus)

```

```python
vendas_diarias = [20,30,40,50]
meta = 30
dias_acima_da_meta = 0
for diaria in vendas_diarias:
  if diaria > meta:
    dias_acima_da_meta = dias_acima_da_meta + 1
print(dias_acima_da_meta)
```

```python
lista_vendas = [1000,500,800,1500,2000,2300]
meta = 1200
contador = 0

for venda in lista_vendas:
  if venda > meta:
    contador += 1
print('Quantidade de vendas que atingiram a meta',contador)
```

```python
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]
meta = 1200
percentual_bonus = 0.1
bonus_extra = 50

for venda in lista_vendas:
  bonus = 0
  if venda >= meta:
    bonus = venda * percentual_bonus
  if venda >= 2000:
    bonus = bonus + bonus_extra
    # bonus += bonus_extra
  print(f'Venda: {venda} -> Bonus : R$ {bonus}')

```

```python
lista_vendas = [1000, 500, 800, 1500, 2000, 2300]

maior_vendas = lista_vendas[0]

for venda in lista_vendas:
    if venda > maior_vendas:
      maior_vendas = venda
print('A maior venda', maior_vendas)

```
